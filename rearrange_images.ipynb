```python
import os
```

```python
church_folders = [x for x in os.listdir("downloads") if os.path.isdir("downloads/" + x)]
for folder in church_folders:
    for file in os.listdir("downloads/" + folder):
        origin = "'downloads/{}/{}'".format(folder, file)
        destination = "'all_images/{}_____{}'".format(folder, file)
        print(destination)
        os.system("cp {} {}".format(origin, destination))
```
